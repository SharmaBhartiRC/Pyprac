```python
# Q. 4 
# accept a charecter and performs 

# Python Program to check character is Alphabet Digit or Special Character
a = input("Please Enter Your Own Character : ")

if((a >= 'a' and a <= 'z') or (a >= 'A' and a <= 'Z')): 
    print("The Given Character ", a, "is an Alphabet") 
elif(a >= '0' and a <= '9'):
    print("The Given Character ", a, "is a numeric Digit")
else:
    print("The Given Character ", a, "is a Special Character")

```
